**Task 1:** Exploring and Visualizing a Simple Dataset

```python
import pandas as pd        # for data handling
import numpy as np
import matplotlib.pyplot as plt       # for plotting graphs
import seaborn as sns          # for advanced visualizations

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, mean_absolute_error, mean_squared_error
```

**Step 1: Load Dataset**

```python
# Loading built-in Iris dataset from seaborn
import seaborn as sns
df = sns.load_dataset('iris')
```

**Step 2: Explore Data**

```python
# Checking shape (rows, columns)
print(df.shape)
```

```python
# Checking column names
print(df.columns)
```

```python
# Displaying first 5 rows
df.head()
```

```python
# Getting detailed info
df.info()
```

**Step 3. SCATTER PLOT**

```python
# Scatter plot to see relationship between sepal_length and petal_length

sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)

plt.title("Sepal vs Petal Length")
plt.show()
```

**Step 4. HISTOGRAM**

```python
# Histogram to see distribution of all numeric features

df.hist(figsize=(10,6))

plt.suptitle("Feature Distribution")
plt.show()
```

**Step 5. BOX PLOT**

```python
# Box plot to detect outliers

sns.boxplot(data=df)

plt.title("Box Plot for Outlier Detection")
plt.show()
```

**CONCLUSION**

```python
# - Petal length is most useful feature
# - Species are clearly separable
# - Dataset is clean with minor outliers
```
